```python
import pandas as pd

# Definir o caminho para o arquivo CSV
csv_file_path = r'D:/OneDrive - InMotion - Consulting/AI Projects/AI-Clinical-Advisory-Crew/DB/consolidated_patient_feedback.csv' 


# Ler o CSV, ignorando linhas problemáticas
df = pd.read_csv(csv_file_path, on_bad_lines='skip')

# Contar as linhas vazias em cada coluna
empty_counts = df.isnull().sum()

# Contar as linhas não vazias em cada coluna
non_empty_counts = df.notnull().sum()

# Criar um DataFrame com a contagem de linhas vazias e não vazias
counts_df = pd.DataFrame({
    'Column': df.columns,
    'Non_Empty_Count': non_empty_counts,
    'Empty_Count': empty_counts
})

# Exibir a tabela com a contagem de linhas vazias e não vazias
print("Contagem de linhas vazias e não vazias em cada coluna:")
print(counts_df)
```
